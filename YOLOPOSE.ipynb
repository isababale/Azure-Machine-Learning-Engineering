```python
from ultralytics import solutions
import cv2
```

```python
cap = cv2.VideoCapture("/content/drive/MyDrive/ISA PHOTOS /isa_pushup.mp4")
assert cap.isOpened(), "Error reading video file"

# get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# initialize video writer to save output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, fps, (frame_width, frame_height))

# initialize AIGymn for squat monitoring
gym = solutions.AIGym(
  show=False,  # Changed show to False for notebook compatibility
  kpts=[11, 13, 15], # Updated kpts to be within the valid range [0, 16]
  model="yolo11x-pose.pt"
)

while cap.isOpened():
    # Read a frame from the video
    success, imo = cap.read()
    if not success:
      print("Video frame is empty or processing is complete")
      break

    results = gym(imo)
    out.write(results.plot_im) # Changed results.annotated_frame to results.plot_im

cap.release()
out.release()
cv2.destroyAllWindows()
```
